```python
import os
import numpy as np
import pickle
import rdkit
from rdkit.Chem import Draw
```

```python
dirs="molecule_ecdf_data"

atoms={}
for f in os.listdir(dirs):
    if f.endswith("_Featurizer.pckl") and f.startswith("GetNumberAtoms"):
        symbol=f.replace("_Featurizer.pckl","").replace("GetNumberAtoms","")
        
        with open(os.path.join(dirs,f),"rb") as _f:
            data=pickle.load(_f)
        atoms[symbol]=data["data"].sum()

del atoms["H"]
atoms_sum=sum([v for v in atoms.values()])
```

```python
atomslist=sorted(list(atoms.keys()),key=lambda k:atoms[k],reverse=True)
rel_atoms={s:atoms[s]/atoms_sum for s in atomslist}
rel_atoms
```

```python
props=np.array(list(rel_atoms.values()))


nprobs=np.zeros_like(props)
nprobs[0]=props[0]
for i in range(1,len(props)):
    fac=1
    for k in range(i):
        fac*=(1-nprobs[k])
    nprobs[i]=props[i]/fac
    

#nprobs[0]=props[0]/fac
nprobs
```

```python
occ=(ra<=nprobs)
cc=occ[occ.any(1)]
print(occ.shape,cc.shape)
```

```python
n_atoms=np.zeros(len(props))
tn_atoms=0

unq, cts = np.unique(cc.argmax(1),return_counts=True)
tn_atoms+=occ.shape[0]
n_atoms[unq]+=cts
```

```python
n_atoms/tn_atoms
```

```python
nprobs=np.abs(nprobs)
atom_rates={atomslist[i]:nprobs[i] for i in range(len(atomslist))}
atom_rates
```

```python
from molNet.utils.mol.generator import generate_n_random_hetero_carbon_lattice
from rdkit import RDLogger
lg = RDLogger.logger()
lg.setLevel(RDLogger.CRITICAL)
    
mols = generate_n_random_hetero_carbon_lattice(100_000,max_c=10,hetero_rates=atom_rates)

#rdkit.Chem.Draw.MolsToGridImage(mols,molsPerRow=5,maxMols=100)
```

```python
test_atoms={s:0 for s in atomslist}
for m in mols:
    for a in m.GetAtoms():
        test_atoms[a.GetSymbol()]+=1
test_atoms_sum=sum([v for v in test_atoms.values()])
test_rel_atoms={s:(test_atoms[s]/test_atoms_sum ,rel_atoms[s] ) for s in atomslist}
test_rel_atoms
```
