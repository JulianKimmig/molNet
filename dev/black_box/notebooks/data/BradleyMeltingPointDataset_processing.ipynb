```python
import pandas as pd
import numpy as np
import sys
from rdkit import Chem
if "../../.." not in sys.path:
    sys.path.append("../../..")
```

```python
raw_df = pd.read_excel("BradleyMeltingPointDataset.xlsx")
raw_df.drop(raw_df[raw_df["donotuse"]=="x"].index,inplace=True)
raw_df.drop(["donotuse","donotusebecause"],axis=1,inplace=True)
raw_df.rename(columns={"smiles":"or_smiles"},inplace=True)
raw_df
```

```python
def norm_smiles(s):
    try:
        return Chem.MolToSmiles(Chem.MolFromSmiles(s))
    except Exception as e:
        return ""
raw_df["smiles"]=raw_df["or_smiles"].apply(norm_smiles)
```

```python
from molNet.utils.identifier2smiles import name_to_smiles
from rdkit.Chem import Descriptors

for idx in  raw_df[raw_df["smiles"]==""]["name"].index:
    name=raw_df.loc[idx,"name"]
    try:
        get_smiles_data=name_to_smiles(name)
    except:
        get_smiles_data = None
    print(name)
    if get_smiles_data is None:
        continue
    
    false_smiles=raw_df.loc[idx,"or_smiles"]
    
    false_mol=Chem.MolFromSmiles(false_smiles,sanitize=False)
    #display(false_mol)
    
    get_smiles=list(get_smiles_data.keys())
    if len(get_smiles)==1:
        true_mol=Chem.MolFromSmiles(get_smiles[0],sanitize=False)
        #print(get_smiles[0])
        if np.abs(Chem.Descriptors.HeavyAtomMolWt(true_mol) - Chem.Descriptors.HeavyAtomMolWt(false_mol))<0.001:
            raw_df.loc[idx,"smiles"]= Chem.MolToSmiles(true_mol)
            #print(idx)
        else:
            print(Chem.Descriptors.HeavyAtomMolWt(true_mol), Chem.Descriptors.HeavyAtomMolWt(false_mol))
            display(true_mol)
            print(get_smiles[0])
            display(false_mol)
            print(false_smiles)
        
    else:
        print(get_smiles)
        break
raw_df.drop(raw_df[raw_df["smiles"]==""].index,inplace=True)
```

```python
raw_df.drop(["link","source","or_smiles","key","csid"],inplace=True,axis=1,errors="ignore")
raw_df
```

```python
raw_df.to_csv("BradleyMeltingPointDataset_processed.csv",index=False)
```

```python
raw_df=pd.read_csv("BradleyMeltingPointDataset_processed.csv")
raw_df.sort_values("smiles", inplace=True)
raw_df
```

```python
#group_desc=group.describe()#.reindex()
```

```python
raw_df["rem"]=-1
#g=
MAX_DT=5
for i in range(20):
    print(i)
    f=0
    for smiles, odata in raw_df[raw_df["rem"]==-1].groupby("smiles"):
        data = odata#[odata["rem"]!=]
        if len(data)==0:
            continue
        if data["mpC"].max()-data["mpC"].min()<=MAX_DT:
            raw_df.loc[data.index,"rem"]=0
            continue

        #temp problem
        f+=1
        if len(data)==2:
            raw_df.loc[data.index,"rem"]=1
            continue

        
        
        
        diff_f_mean = np.abs(data["mpC"]-data["mpC"].mean())
        max_mean_diff_idx=np.argmax(diff_f_mean)
        
        idx2r=data.index[max_mean_diff_idx]
        raw_df.loc[idx2r,"rem"]=1
        data.loc[idx2r,"rem"]=1
        
        # in case single removal is not enough
        if np.abs(data[data["rem"]!=1]["mpC"]-data[data["rem"]!=1]["mpC"].mean()).max() > MAX_DT:
            raw_df.loc[data.index,"rem"]=1
            
    print(f)
    if f==0:
        break
```

```python
raw_df.drop(raw_df[(raw_df["rem"]==1)].index,inplace=True)
raw_df.to_csv("BradleyMeltingPointDataset_processed.csv",index=False)
```

```python
merged_df = raw_df.groupby("smiles").mean()
merged_df = merged_df.reset_index()
merged_df = merged_df.drop("rem",axis=1)

merged_df.to_csv("BradleyMeltingPointDataset_data.csv",index=False)
```
