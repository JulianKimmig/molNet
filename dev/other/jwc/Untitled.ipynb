```python
def factorial(n):
    if n <= 1:
        return 1
    return n + factorial(n-1)
```

```python
#the allmighty numpy package
import numpy as np

#a jupyter specific package, here we load just a specific function to clear the output
from IPython.display import clear_output

#package to work with the operationg system, mostly files and folder
import os

#creates the folder "mb", does nothing if already exists
os.makedirs("mb",exist_ok=True)

#creates an image of the mandelbrot set centered at tcenter_x and center_y
#the zoom level can be any real number
#m and n are the number of pixels (values) of the real and imaginary part
# they should be a power of to to get the least rounding errors due to float point precison
def mandelbrot(center_x=-0.5,center_y=0,zoom=0,m = 2**9,n = 2**9):
    
    #creates linear spaces for real and imaginary part
    x = np.linspace(center_x-2/(2**zoom), center_x+2/(2**zoom), num=m)
    y = np.linspace(center_y+2/(2**zoom), center_y-2/(2**zoom), num=n)
    
    #cretes an grid to have all xy-pairs (real-complex-pairs)
    X,Y = np.meshgrid(x,y)
    
    #creates the actual complex space
    C = X + Y*1j

    
    #counts hof many iterations until limit
    counter=np.zeros_like(C.real)
    #stores which pixels are finished
    done= np.zeros_like(C.real).astype(bool)
    
    #the iteration values of the mandelbrot set
    Z=np.zeros_like(C)
    
    # iteration time depends on zoom level, higher zoom levels need bigger iteration times
    for i in range(int(25*(zoom+1))):
        #calculates the next mandelbot values, but only for the ones NOT done
        Z[~done]=Z[~done]**2 + C[~done]
        #checks which values are done
        done=np.abs(Z)>1000
        #increase the counter for all values that are not done
        counter[~done]+=1
    
    
    #sets the counter for all values which are not done in the end to -1 (this is the actuall mandelbrot set)
    counter[~done]=-1
    
    #plots the 2d image of the counts (eachpices is a specific positon wich an corresponding count) (and sets the axis values to the repective x and y values)
    plt.imshow(counter,extent=[x.min(),x.max(),y.min(),y.max()])
    # adds a bar to show the color-mapped value of the counts 
    plt.colorbar()
    #saves the figure to the mb folder with a name related to the zoom level
    # the image has a resollution of 150 dots per inch (default is 100)
    plt.savefig("mb/mb_{:04.1f}.png".format(zoom),dpi=150)
    #clears previous output
    clear_output(wait=True)
    #shows the figure
    plt.show()
    #closes the figure, that the next plot will create a new image and not plot into the same
    plt.close()

    

#creates an mandelbor imge for the zoomes 0 to 10 with 0.5 stepsize
for i in np.arange(0,10,0.5):
    #zooms to a specific regio
    mandelbrot(
    center_x=-1.3151809804,#-0.7464,
    center_y=0.07348166,#0.1104,
    zoom=i,
    )
    

#package to work with times, like wait or get time
import time

#package to read and write images
from PIL import Image


images=[]
#get all files in the mb folder (sorted by name)
for f in sorted(os.listdir("mb")):
    #only if the filename starts with "mb_"
    if f.startswith("mb_"):
        #opens the image
        i=Image.open("mb/"+f)
        #and adds it to the image list
        images.append(i)

#add the image list in reverse order to itself
images=images+images[::-1]
#endless loop has to be stopped via the stop button
while True:
    #for all images
    for i in images:
        #displays the image
        display(i)
        #wait for 1/24th seconds
        time.sleep(1/24)
        #clear output
        clear_output(wait=True)
    #after full circle wait one second
    time.sleep(1)
    
```
