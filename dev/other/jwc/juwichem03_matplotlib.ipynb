plotting is quite easy in python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,np.pi*5,100)
y=np.sin(x)

plt.plot(x,y)
```

```python
x=np.linspace(0,np.pi*2,100)
for i in range(5):
    y=i*np.sin(x*i)
    plt.plot(x,y,label="{}*sin({}x)".format(i,i))
plt.legend()
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.title("Some sines")
```

### fill area

```python
x=np.linspace(0,10*np.pi,1000)
y=np.sin(x) * np.linspace(1,0,x.shape[0])

delta=x[1]-x[0]

plt.plot(x,y,label="sin")

plt.fill(x,y,alpha=0.5)
plt.legend()

#
plt.plot(x,np.cumsum(y*delta),color="g",label="area",linestyle="--") 

plt.legend()
```
